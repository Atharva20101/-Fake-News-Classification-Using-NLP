#**Text Processing using spacy**

```python
nlp = spacy.load("en_core_web_sm")
```

**Data Cleansing: Removing stopwords and symbols**

```python
def remove_stopwords_and_symbols(text):
    doc = nlp(text)
    filtered_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text
```

```python
x_filtered = []
for i in range(len(x)):
  print(i)
  x_filtered.append(remove_stopwords_and_symbols(x[i]))
```

```python
for i in range(len(x_filtered)):
  x_filtered[i] = x_filtered[i].replace('\n', '')
  x_filtered[i] = x_filtered[i].replace('   ', ' ')
  x_filtered[i] = x_filtered[i].replace('  ', ' ')
```

```python
x_filtered[1]
```

**Tokenization and Lemmatization**

```python
def lemmatize_text(text):
  doc = nlp(text)
  lemmas = []
  for token in doc:
    lemmas.append(token.lemma_)
  return lemmas
```

```python
x_lemma = []
x_lemma = [(print("Position:", i), lemmatize_text(item)) for i, item in enumerate(x_filtered)]
```

```python
x_lemma1 = [tuple(filter(lambda x: x is not None, tup)) for tup in x_lemma]
x_lemma1 = [item for sublist in x_lemma1 for item in sublist]
```

```python
x_lemma1[1]
```

```python
x_final = [[item for item in sublist if not item.lstrip('-').replace('.', '', 1).isdigit()] for sublist in x_lemma1]
```

```python
x_final[1]
```

```python
x_final = [' '.join(sublist) for sublist in x_final]
```

```python
len(x_final)
```

**Splitting data into training and testing set**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final, y, test_size=0.2, random_state=0)
```

```python
print(x_train_c_vectorized.shape[0], len(x_train), x_test_c_vectorized.shape[0], len(x_test))
```

```python
y_pred1 = model1.predict(x_test_c_vectorized)
print(classification_report(y_test, y_pred1))
```

#**GoogleNews Word2Vec**

```python
import gensim
model_word2vec = gensim.models.KeyedVectors.load_word2vec_format("/content/drive/MyDrive/Dataset/GoogleNews-vectors-negative300.bin", binary=True)
```

```python
vocab = model_word2vec.key_to_index
print((len(vocab)))
```

```python
def title_to_vector(title, model):
    words = title.split()
    word_vectors = []
    for word in words:
        if word in model:
            word_vectors.append(model[word])
    if not word_vectors:
        return np.zeros(model.vector_size)
    title_vector = np.mean(word_vectors, axis=0)
    return title_vector
```

```python
x_train_vectors = []
for i, title in enumerate(x_train):
  print(i)
  vector = title_to_vector(title, model_word2vec)
  x_train_vectors.append(vector)
```

```python
print(len(x_train_vectors), len(x_train))
```

```python
x_test_vectors = []
for j, title in enumerate(x_test):
  print(j)
  vector = title_to_vector(title, model_word2vec)
  x_test_vectors.append(vector)
```

```python
print(len(x_test_vectors), len(x_test))
```

```python
LG3 = LogisticRegression(random_state=0)
SVC3 = LinearSVC(random_state=0)
RF3 = RandomForestClassifier(random_state=0)
model7 = LG3.fit(x_train_vectors, y_train)
model8 = SVC3.fit(x_train_vectors, y_train)
model9 = RF3.fit(x_train_vectors, y_train)
```

```python
y_pred7 = model7.predict(x_test_vectors)
print(classification_report(y_test, y_pred7))
```

```python
y_pred8 = model8.predict(x_test_vectors)
print(classification_report(y_test, y_pred8))
```
